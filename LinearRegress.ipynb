### 1.2 - What is linear regression?

- The plot shows these data points: x = [1, 2, 2.5, 3, 4] and y = [1, 4, 7, 9, 15] (red dots).
- Goal: find a straight line y ≈ w * x + b that goes through the cloud of points as closely as possible.
    - w = slope (how steep the line is)  
    - b = intercept (where the line crosses y when x = 0)
- “Best” line means the one that makes the prediction errors small. A common measure is Mean Squared Error (MSE) = average of (y_true − y_pred)^2.
- You can compute w and b with a formula (least squares) or let an optimizer find them (gradient descent).
- Use the fitted line to predict y for new x values.
- Keep in mind: linear regression works well when the relationship is roughly a straight line; it doesn’t handle curved patterns or big outliers well.


```python
x = [1, 2, 2.5, 3, 4]
y = [1, 4, 7, 9, 15]
plt.plot(x, y, 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

##### Line of Best Fit:

```python
plt.plot(x, y, 'ro')
plt.axis([0, 6, 0, 20])

plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
plt.show()
```
